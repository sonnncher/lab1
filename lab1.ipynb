```python
# Завдання №5 ---- int
import csv
import numpy as np
from datetime import datetime

def get_column_values(filename, column_name):
    with open(filename, mode='r', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        header = next(csv_reader)  
        
        if column_name not in header:
            raise ValueError(f"Column '{column_name}' not found in CSV file.")
        
        col_index = header.index(column_name)
        
        column_values = []
        for row in csv_reader:
            value = row[col_index]
            if value:
                try:
                    column_values.append(float(value))
                except ValueError:
                    column_values.append(np.nan)  
            else:
                column_values.append(np.nan)  
        
        return np.array(column_values)

def get_dates(filename):
    with open(filename, mode='r', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        header = next(csv_reader)  
        
        if 'date' not in header:
            raise ValueError("Column 'date' not found in CSV file.")
        
        date_index = header.index('date')
        
        dates = []
        for row in csv_reader:
            date_str = row[date_index]
            if date_str:
                dates.append(datetime.strptime(date_str, '%Y-%m-%d'))  
        
        return np.array(dates)

filename = r'E:\3 курс\Обробка даних Python\лаб 1\archive\russia_losses_equipment.csv'
column_name = 'tank'  
column_values = get_column_values(filename, column_name)

dates = get_dates(filename)

daily_losses = np.diff(column_values) * -1 

daily_losses_clean = daily_losses[~np.isnan(daily_losses)]
dates_clean = dates[1:]  

start_spring = datetime(2023, 3, 1)
end_spring = datetime(2023, 5, 31)

spring_losses = daily_losses_clean[(dates_clean >= start_spring) & (dates_clean <= end_spring)]

total_spring_losses = np.nansum(spring_losses)

print(f"Загальні втрати танків навесні 2023 року: {total_spring_losses}")
```

```python
# Завдання №6
import csv
import numpy as np

def get_column_values(filename, column_name):
    with open(filename, mode='r', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        header = next(csv_reader) 
        
        if column_name not in header:
            raise ValueError(f"Column '{column_name}' not found in CSV file.")
        
        col_index = header.index(column_name)
        
        column_values = []
        for row in csv_reader:
            value = row[col_index]
            if value:
                try:
                    column_values.append(float(value))
                except ValueError:
                    column_values.append(np.nan)  
            else:
                column_values.append(np.nan)  
        
        return np.array(column_values)

filename = r'E:\3 курс\Обробка даних Python\лаб 1\archive\russia_losses_equipment.csv'  
column_name = 'tank' 

column_values = get_column_values(filename, column_name)

daily_losses = np.diff(column_values) * -1 

daily_losses_clean = daily_losses[~np.isnan(daily_losses)]

last_100_days_losses = daily_losses_clean[-100:]

total_losses = np.sum(last_100_days_losses)

average_losses = np.mean(last_100_days_losses)

print(f"Загальні втрати танків за останні 100 днів: {total_losses}")
print(f"Середнє арифметичне втрат танків за останні 100 днів: {average_losses:.2f}")
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd

def get_column_values_and_dates(filename, column_name):
    dates = []
    values = []
    with open(filename, mode='r', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        header = next(csv_reader)  
        
        if column_name not in header:
            raise ValueError(f"Column '{column_name}' not found in CSV file.")
        
        col_index = header.index(column_name)
        
        for row in csv_reader:
            date_str = row[0]  
            value = row[col_index]
            if value:
                try:
                    values.append(float(value))
                    dates.append(datetime.strptime(date_str, '%Y-%m-%d')) 
                except ValueError:
                    values.append(np.nan) 
                    dates.append(np.nan)  
            else:
                values.append(np.nan)  
                dates.append(np.nan)  
        
    return np.array(dates), np.array(values)

filename = r'E:\3 курс\Обробка даних Python\лаб 1\archive\russia_losses_equipment.csv'  
column_name = 'tank'  

dates, column_values = get_column_values_and_dates(filename, column_name)

mask_2023 = (dates >= datetime(2023, 1, 1)) & (dates <= datetime(2023, 12, 31))
dates_2023 = dates[mask_2023]
column_values_2023 = column_values[mask_2023]

daily_losses = np.diff(column_values_2023) * -1  

daily_losses_dates = dates_2023[1:] 
daily_losses_clean = daily_losses[~np.isnan(daily_losses)]

monthly_losses = pd.Series(daily_losses_clean, index=daily_losses_dates[:len(daily_losses_clean)])
monthly_losses = monthly_losses.resample('ME').sum()  

plt.figure(figsize=(12, 6), dpi=100)
plt.plot(monthly_losses.index, monthly_losses.values, linestyle='--', color='blue', marker='o', label='Втрати танків')

plt.grid(color='gray', linestyle=':', linewidth=0.5)

plt.title('Втрати танків за 2023 рік', fontsize=16)
plt.xlabel('Місяць', fontsize=14)
plt.ylabel('Кількість знищених танків', fontsize=14)
plt.legend()

plt.xticks(monthly_losses.index, [date.strftime('%b %Y') for date in monthly_losses.index], rotation=45)

plt.savefig(r'E:\3 курс\Обробка даних Python\лаб 1\tank_losses_2023_monthly.png', bbox_inches='tight')

plt.show()
```
